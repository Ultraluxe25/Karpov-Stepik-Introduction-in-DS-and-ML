````python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
stud_perf = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
stud_perf.head()
````

````python
stud_perf.columns.tolist()
````

````python
stud_perf['math score'].hist();  # выводим гистограмму распределения
````

````python
stud_perf.plot.scatter(x='math score', y='reading score');
````

# [pandas.DataFrame.plot.scatter](https://pandas.pydata.org/pandas-docs/version/0.25.0/reference/api/pandas.DataFrame.plot.scatter.html)

````python
sns.lmplot(x='math score', y='reading score', data=stud_perf);
````

# [seaborn.lmplot](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

атрибут hue - это группирующая переменная

````python
# добавили атрибут hue
sns.lmplot(x='math score', y='reading score', data=stud_perf, hue='gender');
````

````python
# добавили атрибут fit_reg
sns.lmplot(x='math score', y='reading score', data=stud_perf, hue='gender', fit_reg=False);
````

````python
ax = sns.lmplot(x='math score', y='reading score', data=stud_perf, hue='gender', fit_reg=False)
ax.set_xlabels('MATH')
ax.set_ylabels('READING');
````
